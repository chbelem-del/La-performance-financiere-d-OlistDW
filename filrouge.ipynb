```python
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]

for col in date_cols:
    orders[col] = pd.to_datetime(orders[col])
```

```python
La satisfaction globale des clients est élevée, puisque près de 77 % des avis sont des notes de 4 ou 5 étoiles. 
Cependant, une part non négligeable de clients (environ 11,5 %) attribue la note minimale. Il sera intéressant d'analyser 
si cette insatisfaction est liée à des retards de livraison, à certaines catégories de produits ou à certains vendeurs.
```

```python
df = orders.merge(
    payments,
    on="order_id",
    how="left"
)

df = df.merge(
    reviews,
    on="order_id",
    how="left"
)
```

```python
df.shape
```

```python
df.head()
```

```python
#créer la variable: delai de livraison
df["delivery_days"] = (
    df["order_delivered_customer_date"] -
    df["order_purchase_timestamp"]
).dt.days
```

```python
#creer la variable, retard de livraison:

df["late_delivery"] = (
    df["order_delivered_customer_date"] >
    df["order_estimated_delivery_date"]
)
```

```python
#les retards vont ils baisser la satisfaction?:
df.groupby("late_delivery")["review_score"].mean()
```

```python
Les commandes livrées en retard obtiennent une note moyenne de 2,56/5, contre 4,21/5 pour les commandes livrées à temps. 
Les retards ont donc un impact négatif important sur la satisfaction des clients.
```

```python
# Quel est le chiffre d'affaires selon le statut de la commande ?
df.groupby("order_status")["payment_value"].sum()
```

```python
Delivered = 15,49 M
```

```python
#Les commandes livrées plus rapidement obtiennent-elles de meilleures notes ?
df.groupby("review_score")["delivery_days"].mean()
```

```python
On observe une relation inverse entre le délai de livraison et la satisfaction client. 
Les commandes les mieux notées sont livrées plus rapidement que celles recevant de faibles notes.
```

```python
df = orders.merge(payments, on="order_id", how="left")

df = df.merge(reviews, on="order_id", how="left")

df = df.merge(items, on="order_id", how="left")

df = df.merge(products, on="product_id", how="left")

df = df.merge(category,
              left_on="product_category_name",
              right_on="product_category_name",
              how="left")

df = df.merge(customers, on="customer_id", how="left")

df = df.merge(sellers, on="seller_id", how="left")
```

```python
df = df.dropna(subset=["payment_value"])
```

```python
print("===== KPI =====")

print("Chiffre d'affaires :", round(df["payment_value"].sum(),2))

print("Nombre de commandes :", df["order_id"].nunique())

print("Nombre de clients :", df["customer_id"].nunique())

print("Nombre de vendeurs :", df["seller_id"].nunique())

print("Nombre de produits :", df["product_id"].nunique())

print("Note moyenne :", round(df["review_score"].mean(),2))

print("Délai moyen de livraison :", round(df["delivery_days"].mean(),2))

print("Taux de retard :", round(df["late_delivery"].mean()*100,2), "%")
```
